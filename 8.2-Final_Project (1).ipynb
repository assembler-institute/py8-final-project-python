```python
## Loop to calculate the n of blank lines

f.seek(0)
blank_lines = 0
for line in f.readlines():
    if  line == '\n':
        blank_lines += 1

print(f"The file has {blank_lines} blank lines")
```

```python
## Loop to calculate the n of comment lines

f.seek(0)
comment_lines = 0
for line in f.readlines():
    if line.startswith("#") or line.startswith("%") or line.startswith("*"):
        comment_lines +=1

print(f"The file has {comment_lines} commented lines")
```

```python
## valid entries left after cleanup

valid = total_lines - blank_lines - comment_lines
print(f"The number of valid lines is {valid}")
```

```python
## Function that reads a file with raw data and creates a file with clean data

def read_and_clean(f, *args):
    
    """
    This function takes a file path in the form of a string 
    and any number of characters (strings) as arguments. It 
    will create a new file removing all the blank lines and 
    specified characters from the initial file.
    It will return a list with all the valid lines from the 
    initial file.
    
    f = string, file path E.G. "Files/data.txt"
    
    *args = any number of strings E.G. "~", "#", ...
    """

    f = open(f, "r", encoding="utf-8")

# Non-blank lines. Stores not blank entries from data_file in a list

    non_blank_lines = []
    for line in f:
        if line == ("\n"):
            pass
        else:
            non_blank_lines.append(line)
        
# Clean lines with comments
        
    valid_lines = []

    for line in non_blank_lines:
        if line.startswith(args):
            pass
        else:
            valid_lines.append(line)
            
#    return valid_lines

    cleaned_up_data = open("Files/cleaned_up_data.txt", "w", encoding="utf-8")
    
    cleaned_up_data.writelines(valid_lines)
    
    ## prints whether the new file has been created
    
    print("File *cleaned_up_data.txt* has been created") 
    return valid_lines
```

Thus, by setting two repeats as the minimum count parameter, the function provides a dictionary of repeat values, filtering out only entries that repeat more than twice, and reporting how many times each occurs in total.
Specifications: here you will call a function from another function. To produce the counts for each of the entries, you will need to use the `counter()` function previously created. One would be tempted to just write `where_to_look_for.counts(value_to_search)`, but since we want to practice higher-order functions, you will need to implement your `counter(value_to_search, where_to_look_for)` function, which performs the same task but is a user-created function containing higher-order functions and lambda functions.

```python
## Feature picker

def features_picker_cleaner(f, index, char = "", *leftover_chars):
    
    """
    This function takes 3 arguments. An iterable where the filtering 
    will be performed. An index specifying the index number of the 
    object that will be returned and any number of characters that 
    will be removed from said object.
    
    f = an iterable. The function will iterate over this object.
    
    index = an integer. This will be the index of the returned 
    object within the iterable.
    
    char = any number of strings. These characters will be removed 
    from the output string.
    
    """

    for i in leftover_chars:
        char += i
        
## filter() is a built-in function in Python. The filter function can
## be applied to an iterable. It'll create an iterator. This new iterator 
## can filter out certain elements based on the condition provided.
    
    return list(map(lambda x: x.split(",")[index].strip(char), f))
```

```python
## Counter function

def counter(value_to_search, where_to_look_for):
    
    """
    This function takes 2 arguments. One will be the value to be
    searched. The other one will be the object where it'll be 
    searched.
    
    value_to_search = object.
    
    where_to_look_for = an iterable.
    """
    
    return len(list(filter(lambda x: x in value_to_search,  where_to_look_for)))
```

```python
## Values and frequencies

def values_and_frequencies(where_to_look_for, min_num_repetitions):
    
    """
    This function takes 2 arguments, an iterable and the minimum number 
    of instances of any item in that iterable. It returns all the items 
    in the iterable that are repeated the specified number of times in 
    a the form of dictionary.
    
    where_to_look_for = an iterable.
    
    min_num_repetitions = an integer. 
    
    """
           
    return {line:counter(line, where_to_look_for) for line in where_to_look_for if counter(line, where_to_look_for) >= min_num_repetitions}
```

`filter_names()` 

that filters out the second half of the entries’ names, after ordering them alphabetically. Here you will use one of the auxiliary functions defined above: the one that extracts and cleans only the names (first feature) from the entries (which contains three features, remember?).</br>
<br>Parameters: the number and type you consider.</br>
<br>Output: the function should create a text file called "patients_names.txt" with the filtered names. If you want, the function can print some information about feature names when it is executed.</br>


```python
def filter_names(file):
    
    """
    The function takes an iterable. It will remove the duplicates 
    and sort the elements within the iterable alphabetically. Then 
    it returns a list that includes only the second half of the 
    objects in the initial list.
    It will also create a new file with the content of this list, 
    in the location "Files/patient_names.txt"
    
    file = iterable.
    
    """
    
    pacientes = features_picker_cleaner(file, 0, " ", "~", "/", ".")
    setpacientes = set(pacientes)
    pacientes = list(setpacientes)
    pacientes.sort()
    
    pacientes = pacientes[len(pacientes)//2:]
    
    with open ("Files/patient_names.txt", "w", encoding="utf-8") as ju:
        for i in pacientes:
            ju.write(i+"\n")
    
    print("File *patient_names.txt* has been created")
    
    return pacientes
```

<h1>Domain names</h1>
<br>Function to extract the webpage domains. Create a function called 

`domain_extraction()` 

that extracts all non-repeating domains. As you did  for the previous function, fetch the URLs with the features_picker_cleaner() function. Remember that you can use sets to preserve unique values.</br>

<br>Parameters: the number and type you consider.</br>

<br>Return: a list with unique domains and report its structure in the final report</br>


```python
## Function to extract non repeated domain names

def domain_extraction(file):
    
    """
    This function takes an iterable and returns a list with the non 
    duplicates urls in that iterable. It will consider duplicates urls 
    such as "https://www.google.com" and "http://google.com"
    
    file = an iterable.
    """
    
## We use the feature picker function to get the objects with
## index 1 of each line transformed into a list separated by ","
    
    domains = features_picker_cleaner(file, 1, " ", "/", "~", ".")
    
    no_duplicates = []
    
    for url in domains:
        url = url.split("//")[1]
        if url.startswith("www."):
            no_duplicates.append(url.strip("www."))
        else:
            no_duplicates.append(url)
            
    no_duplicates = list(set(no_duplicates))
    
    return no_duplicates
```

```python
## Information function

def characteristics_organization(file):
    
    """
    This function takes an iterable and returns a dictionary with the 
    definitions included in this iterable as key and the number of 
    times each definition is repeated as value.
    
    file = iterable.
    """
    
## Using the features picker function to get the "definition" from the
## clean file
    
    characteristics = features_picker_cleaner(file, 2, " ", "\n", "-", "~")
    
    cleaned_char = []
    
    for i in characteristics:
        if i == "":
            cleaned_char.append("n/a")
        else:
            cleaned_char.append(i)
            
## Using the previously defined counter function to count instances 
## of each definition
            
    dicc = {i:counter(i, cleaned_char) for i in cleaned_char}
    
    return dicc
```
